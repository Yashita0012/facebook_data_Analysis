```python
age_grp = [ '10-20' , '21-30' , '31-40' , '41-50' , '51-60 ' , '61-70' , '71-80' , '81-90' , '91-100', '101 -110','111-120']
df['age_grp']  = pd.cut(df['age'] , bins = np.arange(10,121,10), labels=age_grp, right=True)
df
```

```python
age_grp = df['age_grp'].value_counts().reset_index()
age_grp.sort_values(by= age_grp, inplace=True)
```

```python
plt.figure(figsize=(12,10))
plt.bar( age_grp['age_grp'] , age_grp['count'], color = 'r')
plt.xlabel('DISTRIBUTION OF AGE GROUP' , fontsize = 13)
plt.ylabel('count of peoples' ,fontsize = 13) 
plt.title('AGE GROUP VS COUNT OF PEOPLE ' , fontsize = 13)
plt.show()
```

```python
age_gender_grop = df[['age_grp' , 'gender']].value_counts().sort_values(ascending=False).reset_index()
```

```python
age_gender_grop
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=age_gender_grop, x = 'age_grp' , y= 'count' , hue='gender')
plt.xlabel('age' , fontsize = 12)
plt.ylabel('count', fontsize = 12)
plt.title("age group distribution of both genders" , fontsize = 12)
plt.show()
```
